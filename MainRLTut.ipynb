```python
enviroment_name = 'CartPole-v1'
env = gym.make(enviroment_name, render_mode = 'human')
```

```python
#just the number of times we will run the environment 10 times in this case bc of the for loop
episodes = 10
for episode in range(1, episodes+1):
    #reset resets the environment randomly and gives you the initial set of observations for the env which should be 4 values
    state = env.reset() #these observations will later be passed to our agent to try to maximize reward
    done = False #if the pole angle failed and we finished the trial
    done2 = False #there is a max number of steps or frames in the env and we are tyring to see if we passed it or not
    score = 0

    while not done and not done2:
        env.render() #allows us to view the env
        action = env.action_space.sample() #generate random action b 
        n_state, reward, done, done2, info = env.step(action) # pass the random action into env
        score += reward
    print('Episode:{} Score:{} {} {}'.format(episode, score, done, done2))
#env.close() #close the render frame/window
```

```python
env = gym.make(enviroment_name, render_mode = 'human') #recreated env
env = DummyVecEnv([lambda: env]) #wrapping our enviroment, creating a lambda funciton which is an env creaton fuction,
#its a wrapper for a non vectorized environment 
model = PPO('MlpPolicy', env, verbose=1, tensorboard_log=log_path) #define the model, ppo is the algorithm, define the policy (neural network), 
#we pass the env, verbose=1 to log the results, and then we specify the tensorboard log path
```

# 6. Test Model

```python
env = gym.make(enviroment_name, render_mode = 'human')
```

```python
#just the number of times we will run the environment 10 times in this case bc of the for loop
episodes = 2
means = 0
for episode in range(1, episodes+1):
    #reset resets the environment randomly and gives you the initial set of observations for the env which should be 4 values
    (obs, info) = env.reset() #these observations will later be passed to our agent to try to maximize reward
    done = False #if the pole angle failed and we finished the trial
    done2 = False #there is a max number of steps or frames in the env and we are tyring to see if we passed it or not
    score = 0

    while not done and not done2:
        env.render() #allows us to view the env
        action, _ = model.predict(obs) #NOW IT IS THE MODEL TAKING ACTION
        obs, reward, done, done2, info = env.step(action) # pass the random action into env
        score += reward
    print('Episode: {} Score: {} {} {}'.format(episode, score, done, done2))
    means += score
print('Mean score with {} episodes: {} '.format(episodes, means/episodes))
#env.close() #close the render frame/window
```

# 8. Adding a callback to the training Stage

Callbacks: can stop training or even save a model once a certain requirement is met (Useful for large models so they don't become unstable)

```python
from stable_baselines3.common.callbacks import EvalCallback, StopTrainingOnRewardThreshold
```

```python
save_path = os.path.join('Training', 'Saved Models')
```

```python
stop_callback = StopTrainingOnRewardThreshold(reward_threshold = 250, verbose=1) #stops training once we pass a certain reward
eval_callback = EvalCallback(env, callback_on_new_best=stop_callback, eval_freq=10000, best_model_save_path=save_path, verbose=1) 
#triggered after every run, pass enviroment, the callback to run on new best, how frequent in amount of times per # of timesteps
#best_model saves the model under x
```

```python
model3 = PPO('MlpPolicy', env, verbose=1, tensorboard_log=log_path)
```

```python
model3.learn(total_timesteps=40000, callback=eval_callback)
```
